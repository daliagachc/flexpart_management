```python
def boo(r):
    r0 = xr.DataArray(r.iloc[0])
    r1 = xr.DataArray(r.iloc[1])
    rl_ = flx_ds[co.RL]
    res = np.logical_and(rl_>=r0,rl_<=r1)
    return res

time_stamps['boo']=time_stamps.apply(boo,axis=1)
```
